```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(3, 5)
plt.show()
```

```python
plt.figure()
plt.scatter(3, 5)
plt.show()
```

```python
plt.figure()
plt.subplot(1, 1, 1)
plt.scatter(3, 5)
plt.show()
```

```python
plt.figure()

plt.subplot(1, 2, 1)
plt.scatter(3, 5)

plt.subplot(1, 2, 2)
plt.scatter(3, 5)

plt.show()
```

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.scatter([1, 2, 3, 4, 5, 6], [2, 3, 4, 5, 1, 7])

plt.subplot(2, 1, 2)
plt.plot([1, 2, 3, 4, 5, 6], [2, 3, 4, 5, 1, 7])

plt.show()
```

```python
plt.figure()

x = [9, 6, 4, 3, 6, 2]
y = [4, 7, 2, 3, 7, 3]

plt.subplot(1, 3, 1)
plt.plot(x, y)

plt.subplot(1, 3, 2)
plt.plot(x, y)

plt.subplot(1, 3, 3)
plt.plot(x, y)

plt.show()
```

```python
plt.figure()

x = [9, 6, 4, 3, 6, 2]
y = [4, 7, 2, 3, 7, 3]

for i in range(1, 4):
    plt.subplot(1, 3, i)
    plt.plot(x, y)

plt.show()
```

```python
plt.figure()

x = [9, 6, 4, 3, 6, 2]
y = [4, 7, 2, 3, 7, 3]

for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.plot(x, y)

plt.show()
```

```python
plt.figure()

x = [9, 6, 4, 3, 6, 2]
y = [4, 7, 2, 3, 7, 3]

for i in range(1, 10):
    plt.subplot(3, 3, i)
    plt.plot(x, y)

plt.show()
```
